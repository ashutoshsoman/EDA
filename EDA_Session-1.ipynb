$Strp-3$:
    
**Update the index**

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Name','Age'],
             index=['A','B','C'])
df1
```

```python
df1['Name']=['Sathish','Suresh','Ramesh']
df1
```

$Step-6$:
    
**Add new rows**:  $loc$

```python
# Change the index A B C  to 0  1   2
df1=pd.DataFrame(zip(Names,Age),
             columns=['Name','Age'],
             index=[0,1,2])
df1
```

```python
df1.loc['A']=['Bunny',40]
df1
```

$Step-7$:
    
**Drop the column**

- Before drop the column

- I will create a new column

- I have only two columns are there Name and Age

- I will create a city, Then I will drop that

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Name','Age'],
             index=[0,1,2])
df1['City']=['Hyd','Blr','Chennai']
df1
```

```python
# Axis=1   It reference Column
# Axis=0   It reference rows
# The result You want to overwrite or save in another variable
df1.drop('City',
         axis=1,
        inplace=True)
```

```python
df1
```

$Step-8$

**Drop the row**

```python
df1.drop(2,inplace=True)
```
